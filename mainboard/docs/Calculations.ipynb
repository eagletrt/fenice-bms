# Mainboard calculations

```python
import numpy as np
import matplotlib.pyplot as plt
sqm=lambda x: np.abs(x)**2
```

## Field side ADC filtering in tracking mode

```python
# after the buffer
R1=100e3
C1=100e-9
# inside ADC
R2=500
C2=50e-12

H=lambda w: 1/(R1*R2*C1*C2*(1j*w)**2+R1*C1*(1j*w)+R1*C2*(1j*w)+R2*C2*(1j*w)+1)

w=np.geomspace(1,1000,100)

plt.figure()
plt.suptitle('field ADC filtering')
plt.plot(w,sqm(H(w)))
plt.ylabel('$|H|^2$')
plt.xlabel('$\omega [rad/s]$')
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.axvline(1/(R1*C1))
# plt.axvline(1/(R2*C2))
print('Pole at:',6.28/(R1*C1),'Hz')
```
